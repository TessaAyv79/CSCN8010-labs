```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import fashion_mnist
```

```python
# Load the Fashion MNIST dataset
(train_x, train_y), (test_x, test_y) = fashion_mnist.load_data()
```

```python
# Concatenate train and test data for EDA
all_data = np.concatenate((train_x, test_x))
all_labels = np.concatenate((train_y, test_y))
```

```python
# Check the distribution of classes
class_counts = pd.Series(all_labels).value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize sample images from each class
plt.figure(figsize=(12, 8))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    class_images = all_data[all_labels == i]
    plt.imshow(class_images[0], cmap='gray')
    plt.title(f'Class {i}')
    plt.axis('off')
plt.suptitle('Sample Images from Each Class')
plt.show()
```

```python
# Calculate and visualize pixel-wise mean and standard deviation
pixel_mean = np.mean(all_data, axis=0)
pixel_std = np.std(all_data, axis=0)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(pixel_mean, cmap='gray')
plt.title('Pixel-wise Mean')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(pixel_std, cmap='gray')
plt.title('Pixel-wise Standard Deviation')
plt.axis('off')

plt.suptitle('Pixel-wise Mean and Standard Deviation')
plt.show()
```

```python
# Visualize some random images
plt.figure(figsize=(12, 8))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    rand_idx = np.random.randint(0, len(all_data))
    plt.imshow(all_data[rand_idx], cmap='gray')
    plt.title(f'Class {all_labels[rand_idx]}')
    plt.axis('off')
plt.suptitle('Random Images from the Dataset')
plt.show()
```
